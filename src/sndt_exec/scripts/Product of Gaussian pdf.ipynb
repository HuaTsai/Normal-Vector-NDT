```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from math import sqrt
%matplotlib inline

up, uq = 3, 5
sp, sq = 4, 9
x = np.linspace(-10, 20, 100)
plt.plot(x, stats.norm.pdf(x, up, sqrt(sp)))
plt.plot(x, stats.norm.pdf(x, uq, sqrt(sq)))
plt.plot(x, stats.norm.pdf(x, up, sqrt(sp)) * stats.norm.pdf(x, uq, sqrt(sq)))

cc = stats.norm.pdf(0, uq - up, sqrt(sp + sq))
mc = (up / sp + uq / sq) / (1 / sp + 1 / sq)
sc = 1 / (1 / sp + 1 / sq)
plt.plot(x, cc * stats.norm.pdf(x, mc, sqrt(sc)))

plt.show()
```

```python
print(stats.norm.pdf(0, uq - up, sqrt(sp + sq)))
```

```python
cdf = 0
for i in range(len(x) - 1):
    dx = x[i + 1] - x[i]
    xi1 = stats.norm.pdf(x[i + 1], up, sqrt(sp)) * stats.norm.pdf(x[i + 1], uq, sqrt(sq))
    xi = stats.norm.pdf(x[i], up, sqrt(sp)) * stats.norm.pdf(x[i], uq, sqrt(sq))
    cdf += dx * (xi1 + xi) / 2
print(cdf)
```
