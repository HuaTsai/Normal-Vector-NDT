```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

def heatmap(data, rows, cols, title, bar=True):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.set_title(title)
    im = ax.imshow(data, cmap='RdYlGn', vmin=0)
    if (bar):
        cbar = ax.figure.colorbar(im, ax=ax)
        cbar.ax.set_ylabel('success rate [%]', rotation=-90, va='bottom')

    ax.set_xlabel('[deg]')
    ax.set_xticks(np.arange(data.shape[1]))
    ax.set_xticklabels(cols)
    ax.xaxis.tick_top()

    ax.set_ylabel('[m]')
    ax.set_yticks(np.arange(data.shape[0]))
    ax.set_yticklabels(rows, usetex=True)

    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['top'].set_visible(False)

    ax.set_xticks(np.arange(data.shape[1]+1)-.5, minor=True)
    ax.set_yticks(np.arange(data.shape[0]+1)-.5, minor=True)
    ax.grid(which='minor', color='w', linestyle='-', linewidth=3)
    ax.tick_params(which='minor', bottom=False, left=False)

    fmt = matplotlib.ticker.StrMethodFormatter('{x:.2f}%')
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            im.axes.text(j, i, fmt(data[i, j]), ha='center', va='center', color='k')

radius = [0, 2, 4, 6, 8, 10]
theta = ['$' + str(t) + '^\circ$' for t in [0, 10, 20, 30, 40, 50]]
```

```python
def heatmap2(data, rows, cols, title, bar=True):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.set_title(title)
    im = ax.imshow(data.T, cmap='RdYlGn', vmin=0)
    if (bar):
        cbar = ax.figure.colorbar(im, ax=ax)
        cbar.ax.set_ylabel('success rate [%]', rotation=-90, va='bottom')

    ax.set_xlabel('[m]')
    ax.set_xticks(np.arange(data.shape[0]))
    ax.set_xticklabels(rows, usetex=True)
    ax.xaxis.tick_top()

    ax.set_ylabel('[deg]')
    ax.set_yticks(np.arange(data.shape[1]))
    ax.set_yticklabels(cols)

    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['top'].set_visible(False)

    ax.set_xticks(np.arange(data.shape[0]+1)-.5, minor=True)
    ax.set_yticks(np.arange(data.shape[1]+1)-.5, minor=True)
    ax.grid(which='minor', color='w', linestyle='-', linewidth=3)
    ax.tick_params(which='minor', bottom=False, left=False)

    fmt = matplotlib.ticker.StrMethodFormatter('{x:.2f}%')
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            im.axes.text(i, j, fmt(data[i, j]), ha='center', va='center', color='k')


radius = [0, 3, 6, 9, 12, 15]
theta = ['$' + str(t) + '^\circ$' for t in [0, 20, 40, 60, 80]]
ndt = np.array([
    [100, 100, 100, 65, 30],
    [100, 98.5, 86.5, 57.75, 20.25],
    [60, 57, 40, 21, 8],
    [0, 2.75, 9.25, 3.5, 2],
    [0, 1.25, 1.5, 2, 1],
    [0, 1.25, 1.5, 1, 0]
])
heatmap2(ndt, radius, theta, 'D2D-NDT', False)

nvndt = np.array([
    [100, 100, 95, 90, 80],
    [100, 97, 88.75, 79.75, 67.25],
    [85, 71, 58.75, 53.5, 44.5],
    [30, 33.5, 37.5, 35.25, 31.75],
    [35, 24, 22, 26, 22],
    [10, 15, 15, 16, 13.5]
])
heatmap2(nvndt, radius, theta, 'Normal Vector + NDT')
```

## Contour Representation

```python
def contour(x, y, data, title, bar=True):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.set_title(title)
    cont = ax.contourf(x, y, data, levels=np.arange(101), cmap='RdYlGn')
    if (bar):
        cbar = ax.figure.colorbar(cont, ax=ax, ticks=np.arange(0, 101, 20))
        cbar.ax.set_ylabel('success rate [%]', rotation=-90, va='bottom')
    ax.set_xlabel('[m]')
    ax.set_xticks(np.arange(x[-1] + 1))
    ax.set_ylabel('[deg]')
    ax.set_yticks(np.arange(0, 90, 10))
    ylb = ['$' + str(t) + '^\circ$' for t in np.arange(0, 90, 10)]
    ax.set_yticklabels(ylb)
    ax.set_xlim([0, 10])

radius = [0, 3, 6, 9, 12, 15]
theta = [0, 20, 40, 60, 80]
ndt = np.array([
    [100, 100, 100, 65, 30],
    [100, 98.5, 86.5, 57.75, 20.25],
    [60, 57, 40, 21, 8],
    [0, 2.75, 9.25, 3.5, 2],
    [0, 1.25, 1.5, 2, 1],
    [0, 1.25, 1.5, 1, 0]
])
contour(radius, theta, ndt.T, 'D2D-NDT')

nvndt = np.array([
    [100, 100, 95, 90, 80],
    [100, 97, 88.75, 79.75, 67.25],
    [85, 71, 58.75, 53.5, 44.5],
    [30, 33.5, 37.5, 35.25, 31.75],
    [35, 24, 22, 26, 22],
    [10, 15, 15, 16, 13.5]
])
contour(radius, theta, nvndt.T, 'NV-NDT')
```

```python
def timebar(a, b, labels, title, w=0.35, ndt=True):
    bud = np.array([a[0], b[0]])
    nm  = np.array([a[1], b[1]])
    nd  = np.array([a[2], b[2]])
    opt = np.array([a[3], b[3]])
    oth = np.array([a[4], b[4]])
    fig, ax = plt.subplots(figsize=(3, 4))
    x = [0.5, 1]
    if (ndt):
        ax.bar(labels, opt, w, bottom=nd+bud, color='r', label='Optimize')
        ax.bar(labels, bud, w, bottom=nd, color='c', label='Build')
        ax.bar(labels, nd, w, bottom=0, color='g', label='NDT')
    else:
        ax.bar(labels, opt, w, bottom=nm+bud, color='r', label='Optimize')
        ax.bar(labels, bud, w, bottom=nm, color='c', label='Build')
        ax.bar(labels, nm, w, bottom=0, color='b', label='Normal')
    ax.set_ylabel('[ms]')
    ax.set_title(title)
    ax.legend(loc='upper right')
    return ax
```

```python
def timebar2(a, b, c, d, labels, title, w=0.35, ndt=True):
    bud = np.array([a[0], b[0], c[0], d[0]])
    nm  = np.array([a[1], b[1], c[1], d[1]])
    nd  = np.array([a[2], b[2], c[2], d[2]])
    opt = np.array([a[3], b[3], c[3], d[3]])
    oth = np.array([a[4], b[4], c[4], d[4]])
    fig, ax = plt.subplots(figsize=(4, 4))
    ax.bar(labels, opt, w, bottom=nm+nd+bud, color='r', label='Optimization')
    ax.bar(labels, bud, w, bottom=nm+nd, color='c', label='Association')
    ax.bar(labels, nd, w, bottom=nm, color='g', label='NDT Models')
    ax.bar(labels, nm, w, bottom=0, color='b', label='Normal Vectors')
    ax.set_ylabel('[ms]')
    ax.set_title(title)
    ax.legend(loc='upper right')
    
    axins = ax.inset_axes((1.2, 0.1, 0.5, 0.7))
    axins.bar(labels, opt, w, bottom=nm+nd+bud, color='r', label='Optimize')
    axins.bar(labels, bud, w, bottom=nm+nd, color='c', label='Build')
    axins.bar(labels, nd, w, bottom=nm, color='g', label='NDT')
    axins.bar(labels, nm, w, bottom=0, color='b', label='Normal')
    axins.set_xlim(1.6, 3.4)
    axins.set_ylim(0, 180)

    ax.indicate_inset_zoom(axins, edgecolor="black")
    return ax
```

```python
icp = [1997.88, 0.00, 0.00, 855.56, 12.22]
sicp = [614.33, 789.20, 0.00, 313.29, 1.67]
ndt = [27.31, 0.00, 12.63, 112.09, 0.11]
nvndt = [16.41, 0.00, 12.58, 67.03, 0.08]
ax = timebar2(icp, sicp, ndt, nvndt, ['ICP', 'Symm-ICP', 'D2D-NDT', 'NV-NDT'], 'Used Time', 0.3)
```
