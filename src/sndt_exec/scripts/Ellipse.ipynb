```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

xc, yc = 0, 0
x, y = 10, 6
xs, ys = [], []
for i in range(360):
    th = np.deg2rad(i)
    xs.append(xc + x * np.cos(th))
    ys.append(yc + y * np.sin(th))

fig, ax = plt.subplots(figsize=(16, 9))
ax.set_xticks(list(range(-100, 100)))
ax.set_yticks(list(range(-100, 100)))
ax.grid()
ax.scatter(xs, ys)
```
