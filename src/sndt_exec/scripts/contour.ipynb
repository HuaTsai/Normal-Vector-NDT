```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.colors import LogNorm
from IPython.display import HTML
from math import sqrt
from math import pi
%matplotlib inline
```

```python
def Contour():
    with open('/tmp/contour.txt', 'r') as file:
        line = len(file.readlines())
    x = np.genfromtxt('/tmp/contour.txt', delimiter=',', skip_footer=line-1)
    y = np.genfromtxt('/tmp/contour.txt', delimiter=',', skip_header=1, skip_footer=line-2)
    zz = np.genfromtxt('/tmp/contour.txt', delimiter=',', skip_header=2)
    print(zz[20, 20])
    xx, yy = np.meshgrid(x, y)
    gt = [-15, -15]
    gt = [-14.4966, 3.85344]
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(*gt, 'r*', markersize=16, label='Ground Truth')
    ax.plot(0, 0, 'bX', markersize=16, label='Initial Pose')
    ax.set_xlabel('x [m]')
    ax.set_ylabel('y [m]')
    ax.contour(xx, yy, zz, levels=10, cmap=plt.cm.jet)
    ax.grid()
    ax.legend()
```

```python
Contour()
```
