```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.colors import LogNorm
from IPython.display import HTML
from math import sqrt
from math import pi
%matplotlib inline

%matplotlib
fig, ax = plt.subplots(figsize=(10, 6))
xx, tt, zz, cont, gt, pos = None, None, None, None, None, None
%matplotlib inline

def update(i):
    ax.cla()
    ax.set_xlabel('x [m]')
    ax.set_ylabel('t [deg]')
    ax.grid()
    ax.plot(gt[0], gt[2], 'r*', markersize=16, label='Ground Truth')
    cont = ax.contour(xx, tt, zz[i], levels=15, cmap=plt.cm.jet)
    ax.clabel(cont, fmt='%2.1f', use_clabeltext=False)
    ax.plot([p[0] for p in pos[:i+1]], [p[2] for p in pos[:i+1]], 'mo-', label='Optimize Path')
    ax.legend()

def ContourXT():
    global xx, tt, zz, gt, pos
    f = '/tmp/contourxt.txt'
    with open(f, 'r') as file:
        n = len(file.readlines())
    gt = np.genfromtxt(f, delimiter=',', skip_footer=n-1)
    x = np.genfromtxt(f, delimiter=',', skip_header=1, skip_footer=n-2)
    t = np.genfromtxt(f, delimiter=',', skip_header=2, skip_footer=n-3)
    xx, tt = np.meshgrid(x, t)
    zz, pos = [], []
    sz = t.size + 1
    sh, sf = 3, n - 3 - sz
    fs = int((n - 3) / sz)
    while (sh + sz <= n):
        pos.append(np.genfromtxt(f, delimiter=',', skip_header=sh, skip_footer=sf+sz-1))
        zz.append(np.genfromtxt(f, delimiter=',', skip_header=sh+1, skip_footer=sf))
        sh += sz
        sf -= sz
    ani = FuncAnimation(fig, update, frames=fs, interval=1000, repeat_delay=5)
    return HTML(ani.to_html5_video())
```

```python
ContourXT()  # Symmetric ICP
```
