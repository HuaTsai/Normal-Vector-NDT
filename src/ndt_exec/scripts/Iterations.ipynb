```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
%matplotlib inline

def plot(ax, x, it, t):
    if (t == 'icp'):
        ax.plot(x, it[0,:], label='icp')
#     ax.plot(x, it[1,:], label='point-to-plane icp')
        ax.plot(x, it[2,:], label='symmetric icp')
    if (t == 'ndt'):
        ax.plot(x, it[3,:], label='ndt')
        ax.plot(x, it[4,:], label='symmetric ndt')
    ax.legend()

def func(it, t):
    fig = plt.figure(figsize=(20, 20))
    x = range(it.shape[1])
    ax = fig.add_subplot(511)
    plot(ax, x[:600], it[:,:600], t)
    ax = fig.add_subplot(512)
    plot(ax, x[600:1200], it[:,600:1200], t)
    ax = fig.add_subplot(513)
    plot(ax, x[1200:1800], it[:,1200:1800], t)
    ax = fig.add_subplot(514)
    plot(ax, x[1800:2400], it[:,1800:2400], t)
    ax = fig.add_subplot(515)
    plot(ax, x[2400:3000], it[:,2400:3000], t)
```

```python
it = np.genfromtxt('./iterations.txt', delimiter=',', skip_footer=30)
func(it, 'icp')
```

```python
func(it, 'ndt')
```
